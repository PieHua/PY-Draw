```python
import pandas as pd

# Replace 'G:\Temp\salts-1.xlsx' with your Excel file's path
file_path = 'G:\Temp\salts-1.xlsx'

# Read Excel data into a DataFrame
df = pd.read_excel(file_path)

# Create a dictionary to store the results
results = {'Time': [], 'Interface_Position': []}

# Iterate through concentration columns (starting from the second column)
for col in df.columns[1:]:
    # Find the first row where concentration is less than or equal to 1
    negative_rows = df[df[col] <= 1].index
    
    if len(negative_rows) > 0:
        first_negative_row = negative_rows[0]
        # Get the corresponding position value from the first column
        position_at_threshold = df.loc[first_negative_row, 'position']
        results['Time'].append(col)
        results['Interface_Position'].append(position_at_threshold)
    else:
        results['Time'].append(col)
        results['Interface_Position'].append(None)

# Create a new DataFrame from the results
results_df = pd.DataFrame(results)

# Specify the path for the new Excel file
output_file_path = 'interface_positions.xlsx'

# Save the results to a new Excel file
results_df.to_excel(output_file_path, index=False)

print(f"Results saved to '{output_file_path}'.")
```
