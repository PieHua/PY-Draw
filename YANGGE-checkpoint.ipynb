```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data from the Excel file
excel_file = r'G:\Temp\1.xlsx'
df = pd.read_excel(excel_file)

# Define the columns for the x-axis (time) and y-axes ('NH4', 'NO2', 'NO3', 'DON')
x_column = 'time'
y_columns = ['NH4', 'NO2', 'NO3', 'DON']

# Create a Seaborn line plot
sns.set(style="darkgrid")  # Set the style
plt.figure(figsize=(10, 6))  # Set the figure size

for col in y_columns:
    sns.lineplot(x=x_column, y=col, data=df, label=col)

plt.title("Time Series Data")  # Set the title
plt.xlabel("Time")  # X-axis label
plt.ylabel("Concentration")  # Y-axis label
plt.legend()  # Display the legend

# Show the plot
plt.show()
```
