```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from Excel file
file_path = r'G:\Temp\salt.data.xlsx'
reclaimed_area1_df = pd.read_excel(file_path, sheet_name='Reclaimed Area 1')
reclaimed_area2_df = pd.read_excel(file_path, sheet_name='Reclaimed Area 2')
reclaimed_area3_df = pd.read_excel(file_path, sheet_name='Reclaimed Area 3')

# Assuming the columns are labeled as "Time," "Total Salt Mass," and "Wedge Length"
time_points = reclaimed_area1_df['Time']
reclaimed_area1_salt_mass1 = reclaimed_area1_df['Total Salt Mass1']
reclaimed_area2_salt_mass1 = reclaimed_area2_df['Total Salt Mass1']
reclaimed_area3_salt_mass1 = reclaimed_area3_df['Total Salt Mass1']

reclaimed_area1_salt_mass2 = reclaimed_area1_df['Total Salt Mass2']
reclaimed_area2_salt_mass2 = reclaimed_area2_df['Total Salt Mass2']
reclaimed_area3_salt_mass2 = reclaimed_area3_df['Total Salt Mass2']

reclaimed_area1_wedge_length = reclaimed_area1_df['Wedge Length']
reclaimed_area2_wedge_length = reclaimed_area2_df['Wedge Length']
reclaimed_area3_wedge_length = reclaimed_area3_df['Wedge Length']

# Create subplots to display both salt mass and wedge length
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 13), sharex=True)

# Plot Total Salt Mass for each area
ax1.plot(time_points, reclaimed_area1_salt_mass1, label='Reclaimed Area 1')
ax1.plot(time_points, reclaimed_area2_salt_mass1, label='Reclaimed Area 2')
ax1.plot(time_points, reclaimed_area3_salt_mass1, label='Reclaimed Area 3')
ax1.set_ylabel('Total Salt Mass1')
ax1.set_title('Salt Mass1 Trends in Different Reclaimed Areas')
ax1.legend()

# Plot Wedge Length for each area
ax2.plot(time_points, reclaimed_area1_wedge_length, label='Reclaimed Area 1')
ax2.plot(time_points, reclaimed_area2_wedge_length, label='Reclaimed Area 2')
ax2.plot(time_points, reclaimed_area3_wedge_length, label='Reclaimed Area 3')
ax2.set_xlabel('Time')
ax2.set_ylabel('Wedge Length')
ax2.set_title('Wedge Length Trends in Different Reclaimed Areas')
ax2.legend()

ax3.plot(time_points, reclaimed_area1_salt_mass2, label='Reclaimed Area 1')
ax3.plot(time_points, reclaimed_area2_salt_mass2, label='Reclaimed Area 2')
ax3.plot(time_points, reclaimed_area3_salt_mass2, label='Reclaimed Area 3')
ax3.set_ylabel('Total Salt Mass2')
ax3.set_title('Salt Mass2 Trends in Different Reclaimed Areas')
ax3.legend()

# Adjust layout
plt.tight_layout()

# Save the graph as an image file
plt.savefig('salt_mass_wedge_length.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.savefig('SMT&SWL.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import savgol_filter

# Read data from Excel file
file_path = r'G:\Temp\salt.data.xlsx'
reclaimed_area1_df = pd.read_excel(file_path, sheet_name='Reclaimed Area 1')
reclaimed_area2_df = pd.read_excel(file_path, sheet_name='Reclaimed Area 2')
reclaimed_area3_df = pd.read_excel(file_path, sheet_name='Reclaimed Area 3')

# Assuming the columns are labeled as "Time" and "Wedge Length"
time_points = reclaimed_area1_df['Time']
reclaimed_area1_wedge_length = reclaimed_area1_df['Wedge Length']
reclaimed_area2_wedge_length = reclaimed_area2_df['Wedge Length']
reclaimed_area3_wedge_length = reclaimed_area3_df['Wedge Length']

# Apply Savitzky-Golay filter for smoothing
window_size = 11  # Adjust the window size as needed
degree = 3  # Polynomial degree
smoothed_area1_wedge_length = savgol_filter(reclaimed_area1_wedge_length, window_size, degree)
smoothed_area2_wedge_length = savgol_filter(reclaimed_area2_wedge_length, window_size, degree)
smoothed_area3_wedge_length = savgol_filter(reclaimed_area3_wedge_length, window_size, degree)

# Create subplots to display smoothed wedge length
fig, ax = plt.subplots(figsize=(10, 6))

# Plot Smoothed Wedge Length for each area
ax.plot(time_points, smoothed_area1_wedge_length, label='Reclaimed Area 1')
ax.plot(time_points, smoothed_area2_wedge_length, label='Reclaimed Area 2')
ax.plot(time_points, smoothed_area3_wedge_length, label='Reclaimed Area 3')
ax.set_xlabel('Time')
ax.set_ylabel('Smoothed Wedge Length')
ax.set_title('Smoothed Wedge Length Trends in Different Reclaimed Areas')
ax.legend()

# Adjust layout
plt.tight_layout()

# Save the graph as an image file
plt.savefig('salt_mass_smoothed_wedge_length.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from Excel file
file_path = r'G:\Temp\salt.data.xlsx'
reclaimed_area1_df = pd.read_excel(file_path, sheet_name='Reclaimed Area 1')
reclaimed_area2_df = pd.read_excel(file_path, sheet_name='Reclaimed Area 2')
reclaimed_area3_df = pd.read_excel(file_path, sheet_name='Reclaimed Area 3')

# Assuming the columns are labeled as "Time" and "Wedge Length"
time_points = reclaimed_area1_df['Time']
reclaimed_area1_wedge_length = reclaimed_area1_df['Wedge Length']
reclaimed_area2_wedge_length = reclaimed_area2_df['Wedge Length']
reclaimed_area3_wedge_length = reclaimed_area3_df['Wedge Length']

# Create subplots to display step plot of wedge length
fig, ax = plt.subplots(figsize=(10, 6))

# Create step plots for each area
ax.step(time_points, reclaimed_area1_wedge_length, label='Reclaimed Area 1', where='mid')
ax.step(time_points, reclaimed_area2_wedge_length, label='Reclaimed Area 2', where='mid')
ax.step(time_points, reclaimed_area3_wedge_length, label='Reclaimed Area 3', where='mid')

ax.set_xlabel('Time')
ax.set_ylabel('Wedge Length')
ax.set_title('Step Plot of Wedge Length Trends in Different Reclaimed Areas')
ax.legend()

# Adjust layout
plt.tight_layout()

# Save the graph as an image file
plt.savefig('salt_mass_step_wedge_length.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```
